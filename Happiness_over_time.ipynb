Getting the mean values of happiness rank in each year for each region

```
df_happiness_WesternEurope =df_integrated_happinies[df_integrated_happinies.Region == "Western Europe"]
df_happiness_mean_WesternEurope= df_happiness_WesternEurope.groupby(['Year']).mean()
df_happiness_NorthAmerica =df_integrated_happinies[df_integrated_happinies.Region == "North America"]
df_happiness_mean_NorthAmerica= df_happiness_NorthAmerica .groupby(['Year']).mean()
df_happiness_Australia  =df_integrated_happinies[df_integrated_happinies.Region == "Australia and New Zealand"]
df_happiness_mean_Australia = df_happiness_Australia.groupby(['Year']).mean()
df_happiness_MiddleEast =df_integrated_happinies[df_integrated_happinies.Region == "Middle East and Northern Africa"]
df_happiness_mean_MiddleEast= df_happiness_MiddleEast .groupby(['Year']).mean()
df_happiness_LatinAmerica  =df_integrated_happinies[df_integrated_happinies.Region == "Latin America and Caribbean"]
df_happiness_mean_LatinAmerica = df_happiness_LatinAmerica.groupby(['Year']).mean()
df_happiness_Asia=df_integrated_happinies[df_integrated_happinies.Region == "Southeastern Asia"]
df_happiness_mean_Asia= df_happiness_Asia.groupby(['Year']).mean()
df_happiness_Europe =df_integrated_happinies[df_integrated_happinies.Region == "Central and Eastern Europe"]
df_happiness_mean_Europe= df_happiness_Europe.groupby(['Year']).mean()
df_happiness_EasternAsia =df_integrated_happinies[df_integrated_happinies.Region == "Eastern Asia"]
df_happiness_mean_EasternAsia=df_happiness_EasternAsia.groupby(['Year']).mean()
df_happiness_SubAfrica =df_integrated_happinies[df_integrated_happinies.Region == "Sub-Saharan Africa"]
df_happiness_mean_SubAfrica= df_happiness_SubAfrica.groupby(['Year']).mean()
df_happiness_SouthAsia =df_integrated_happinies[df_integrated_happinies.Region == "Southern Asia"]
df_happiness_mean_SouthAsia= df_happiness_SouthAsia.groupby(['Year']).mean()
```

plotting the change of The mean of the mean of the happiness score for each Region over the Years

```
import matplotlib.pyplot as plt
plt.figure(figsize=(25,10))
#fig_1 = plt.figure(figsize=(6,6))
#ax_1 = fig_1.add_axes([0.1,0.1,0.9,0.9]) #left, bottom, width, and height
#plt.set_xlabel('Year')
#plt.set_ylabel('Happiness score')
#y=X.sort_values(by = "Year")
plt.plot(np.array([2015,2016,2017,2018,2019]),np.array(df_happiness_mean_WesternEurope["Happiness Score"]),label = "Western Europe")
plt.plot(np.array([2015,2016,2017,2018,2019]),np.array(df_happiness_mean_NorthAmerica["Happiness Score"]),label = "North America")
plt.plot(np.array([2015,2016,2017,2018,2019]),np.array(df_happiness_mean_Australia["Happiness Score"]),label = "Australia and New Zealand")
plt.plot(np.array([2015,2016,2017,2018,2019]),np.array(df_happiness_mean_MiddleEast["Happiness Score"]),label = "Middle East and Northern Africa")
plt.plot(np.array([2015,2016,2017,2018,2019]),np.array(df_happiness_mean_LatinAmerica["Happiness Score"]),label = "Latin America and Caribbean")
plt.plot(np.array([2015,2016,2017,2018,2019]),np.array(df_happiness_mean_Asia["Happiness Score"]),label = "Southeastern Asia")
plt.plot(np.array([2015,2016,2017,2018,2019]),np.array(df_happiness_mean_Europe["Happiness Score"]),label = "Central and Eastern Europe")
plt.plot(np.array([2015,2016,2017,2018,2019]),np.array(df_happiness_mean_EasternAsia["Happiness Score"]),label = "Eastern Asia")
plt.plot(np.array([2015,2016,2017,2018,2019]),np.array(df_happiness_mean_SubAfrica["Happiness Score"]),label = "Sub-Saharan Africa")
plt.plot(np.array([2015,2016,2017,2018,2019]),np.array(df_happiness_mean_SouthAsia["Happiness Score"]),label = "Southern Asia")
plt.legend()
# Display a figure.
plt.show()
#ax_1.set_xscale('log')

```

As we see in the above graph which illustrates the change of The happiness Score over the Years for each Region. 
We can see thet the ranking of the regions of the happiness score from the happiest to the less happier: 
australia and newzeland, North America, Western Europe, Latin America, SouthEastern Asia,Middle east and africa, Central and Eastern Europe,  SouthEastern Asia, Southern Asia, Sub Saharan Africa. 
For the Western Europe Region, we can see that, the happiness score was around 6.6 for the period from 2015 - 2017 and it increaed  slightly for around 6.8 from 2017 till 2019. 
For the North America Region, we can see that, the happiness score was around 7.5 for the period from 2015 - 2016 and it decreased slightly for around 7.3 from 2016 till 2019.
For the Australia and North America Region, we can see that, the happiness score was almost fixed over the 5 years 
For the Latin America Region, we can see that, the happiness score was around 6.1 for the period from 2015 - 2016 and it decreased slightly for around 5.9 from 2016 till 2019. 
For the SouthEastern Region, we can see that, the happiness score was increasing during the period 2015-2017, but it started to decrease from 2017-2019 
For the Central and Eastern Europe Region, we can see that, the happiness score was increasing from 5.4 to 5.65 during the period 2015-2019
For the Eastern Asia Region, we can see that, the happiness score was decreasing 2015-2017 during the period 2015-2017, but it started to increase from 2017-2019

For the Sub-saharan Region, we can see that, the happiness score was decreasing 2015-2017 during the period 2015-2017, but it started to increase from 2017-2019

For the Southern Asia Region, we can see that, the happiness score was decreasing 2015-2017 during the period 2015-2017, but it started to increase from 2017-2019

In conclusion, we found out that there's no significant change in the happiness score over the years for each Region, the happiness score can increase or decrease slightly.
